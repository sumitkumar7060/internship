```python
import selenium
import numpy as np
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
# Q1 .
driver=webdriver.Chrome()
driver.get('https://www.naukri.com/')
try:
    title=driver.find_element(By.CLASS_NAME,'suggestor-input ')
    title.send_keys('Data Scientist')
    search=driver.find_element(By.XPATH,'/html/body/div[1]/div[7]/div/div/div[6]')
    search.click()
    time.sleep(3)
    loc_delhi = driver.find_element(By.ID, 'chk-Delhi / NCR-cityTypeGid-')
    driver.execute_script("arguments[0].click();", loc_delhi)
    time.sleep(5)
    salary = driver.find_element(By.ID, 'chk-3-6 Lakhs-ctcFilter-')
    driver.execute_script("arguments[0].click();", salary)
    time.sleep(3)

    job_title=[]
    job_loc=[]
    company=[]
    experience=[]
    titles = WebDriverWait(driver, 30).until(EC.visibility_of_all_elements_located((By.XPATH, '//div[@class="cust-job-tuple layout-wrapper lay-2 sjw__tuple "]/div/a')))
    for t in titles[0:10]:
        job_title.append(t.text)
    loc=driver.find_elements(By.XPATH,'//span[@class="locWdth"]')
    for l in loc[0:10]:
        job_loc.append(l.text)
    comp=driver.find_elements(By.XPATH,'//span[@class=" comp-dtls-wrap"]/a[1]')
    for c in comp[0:10]:
        company.append(c.text)
    exp=driver.find_elements(By.XPATH,'//span[@class="expwdth"]')
    for e in exp[0:10]:
        experience.append(e.text)
    df=pd.DataFrame({'Title':job_title,'Comapny Name':company,'Location':job_loc,'Experience':experience})
    print(df)

except Exception as e:
    print('some thing else',e)
```

```python
# Q2
try:
    driver=webdriver.Chrome()
    driver.get('https://www.shine.com/')
    designation = WebDriverWait(driver, 60).until(EC.visibility_of_element_located((By.CLASS_NAME, "form-control")))
    designation.send_keys('Data Analyst')
    location = WebDriverWait(driver, 60).until(EC.visibility_of_element_located((By.XPATH, "/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[2]/div/input")))
    location.send_keys('Bangalore')
    search=driver.find_element(By.XPATH,"/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[2]/div/button")
    search.click()

    job_title=[]
    job_location=[]
    company_name=[]
    experience_required=[]
    title=driver.find_elements(By.XPATH,'//strong[@class="jobCard_pReplaceH2__xWmHg"]/h2/a')
    for i in title[0:10]:
        job_title.append(i.text)
    loc=driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')
    for l in loc[0:10]:
        loc_text=l.text.split('\n')
        job_location.append(loc_text[0])

    comp=driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]/span')
    for c in comp[0:10]:
        company_name.append(c.text)
    exp=driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')
    for e in exp[0:10]:
        experience_required.append(e.text)

    df=pd.DataFrame({'Title':job_title,'Location':job_location,'Company':company_name,'Experience':experience_required})
    print(df)
    
except TimeoutException:
    print("Element not found within the specified timeout.")


```

```python
# Q3.
try:
    driver=webdriver.Chrome()
    driver.get('https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART')
    rating=[]
    review=[]
    full_review=[]
    c=11
    i=1
    while True:
        rate=driver.find_elements(By.XPATH,'//div[@class="XQDdHH Ga3i8K"]')
        for r in rate:
            rating.append(r.text)
        rev=driver.find_elements(By.XPATH,'//p[@class="z9E0IG"]')
        for rv in rev:
            review.append(rv.text)
        frev=driver.find_elements(By.XPATH,'//div[@class="ZmyHeo"]/div/div')
        for fr in frev:
            full_review.append(fr.text)
        next=driver.find_element(By.XPATH,f'/html/body/div/div/div[3]/div/div/div[2]/div[13]/div/div/nav/a[{c}]/span')
        next.click()
        time.sleep(3)
        c=12
        if(len(rating)==100):
            break
    df=pd.DataFrame({'Rating':rating,'Review':review,'Full_Review':full_review})
    print(df)
    
except Exception as e:
    print('some thing else',e)
```

```python
# Q4.
try:
    driver=webdriver.Chrome()
    driver.get('https://www.flipkart.com/')
    product=driver.find_element(By.CLASS_NAME,'Pke_EE')
    product.send_keys('sneakers')
    search=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div/div/div[1]/div[1]/header/div[1]/div[2]/form/div/button')
    search.click()
    time.sleep(3)
    brand=[]
    desc=[]
    price=[]
    c=11

    for i in range(0,3):
        
        Brand=driver.find_elements(By.XPATH,'//div[@class="syl9yP"]')
        descr=driver.find_elements(By.XPATH,'//div[@class="hCKiGj"]/a[1]')
        pr=driver.find_elements(By.XPATH,'//div[@class="Nx9bqj"]')
        if(i==2):
            for b,d,p in zip(Brand[0:20],descr[0:20],pr[0:20]):
                brand.append(b.text)
                desc.append(d.text)
                price.append(p.text)
        else:
            for b,d,p in zip(Brand,descr,pr):
                brand.append(b.text)
                desc.append(d.text)
                price.append(p.text)
    
        next=driver.find_element(By.XPATH,f'/html/body/div/div/div[3]/div[1]/div[2]/div[12]/div/div/nav/a[{c}]')
        next.click()
        c=12
        time.sleep(3)
    df=pd.DataFrame({'Brand':brand,'Description':desc,'Price':price})
    print(df)
except Exception as e:
    print('something else',e)
```

```python
# Q5.
try:
    driver=webdriver.Chrome()
    driver.get('https://www.amazon.in/')
    pro=driver.find_element(By.XPATH,'/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input')
    pro.send_keys('Laptop')
    search=driver.find_element(By.XPATH,'/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span/input')
    search.click()
    time.sleep(3)
    filter = driver.find_element(By.XPATH, "//span[contains(@data-csa-c-content-id,'p_n_feature_thirteen_browse-bin/12598163031')]//input[@type='checkbox']")
    driver.execute_script("arguments[0].click();", filter)
    time.sleep(3)
    title=[]
    rating=[]
    price=[]
    tl=driver.find_elements(By.XPATH,'//span[@class="a-size-medium a-color-base a-text-normal"]')
    for t in tl[0:10]:
        title.append(t.text)
    rate=driver.find_elements(By.XPATH,'//div[@class="a-row a-size-small"]/span[2]/a/span')
    for r in rate[0:10]:
        rating.append(r.text)
    pr=driver.find_elements(By.XPATH,'//span[@class="a-price-whole"]')
    for p in pr[0:10]:
        price.append(p.text)
    df=pd.DataFrame({'Title':title,'Rating':rating,'Price':price})
    print(df.head(10))

except Exception as e:
    print('some error occur',e)
```

```python
# Q6
try:
    driver=webdriver.Chrome()
    driver.get('https://www.azquotes.com/')
    top=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')
    top.click()
    quotes=[]
    authors=[]
    type_quotes=[]
    c=12
    for i in range(10):
        quote=driver.find_elements(By.XPATH,'//a[@class="title"]')
        for q in quote:
            quotes.append(q.text)
    
        author=driver.find_elements(By.XPATH,'//div[@class="author"]/a')
        for a in author:
            authors.append(a.text)
        type_quote=driver.find_elements(By.XPATH,'//div[@class="tags"]')
        for t in type_quote:
            type_quotes.append(t.text)
        if(i>8):
            break
        next=driver.find_element(By.XPATH,f'/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[3]/li[{c}]/a')
        next.click()
        c=13
        time.sleep(2)
    for qq,au,ty in zip(quotes,authors,type_quotes):
        print('Quotes :',qq)
        print('Author :',au)
        print('Type Quotes :',ty)

except Exception as e:
    print('some thing else',e)
```

```python
# Q7
try:
    driver=webdriver.Chrome()
    driver.get('https://www.jagranjosh.com/general-knowledge/list-of-all-prime-ministers-of-india-1473165149-1')

    Name=[]
    Born_Dead=[]
    Term=[]
    Remarks=[]
    table=driver.find_element(By.XPATH,'//div[@class="TableData"]/table')
    rows = table.find_elements(By.TAG_NAME, 'tr')
    for row in rows[1:]:
        data = row.find_elements(By.TAG_NAME, 'td')
        Name.append(data[1].text)
        Born_Dead.append(data[2].text)
        Term.append(data[3].text)
        Remarks.append(data[4].text)

    df=pd.DataFrame({'Name':Name,'Born-Dead':Born_Dead,'Term Of Office':Term,'Remarks':Remarks})
    print(df)

except Exception as e:
    print('some error occured',e)
```
