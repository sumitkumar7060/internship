```python
# Q3
'''Write a python program to access the search bar and search button on images.google.com.
scrape 10 images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’, ‘Guitar’, ‘Cakes’.'''
def search_img(product):
    try:
        driver_3 = webdriver.Chrome()
        driver_3.get('https://images.google.com/')
        
        search_bar = WebDriverWait(driver_3, 30).until(
            EC.visibility_of_element_located((By.XPATH, "/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/textarea"))
        )
        search_bar.send_keys(product)
        
        button_3 = WebDriverWait(driver_3, 5).until(
            EC.element_to_be_clickable((By.XPATH, "/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/button"))
        )
        driver_3.execute_script("arguments[0].click();", button_3)
        
        time.sleep(5)
        
        for _ in range(20):
            driver_3.execute_script("window.scrollBy(0,500)")
        
        imgs = WebDriverWait(driver_3, 10).until(
            EC.presence_of_all_elements_located((By.XPATH, '//img[@class="YQ4gaf"]'))
        )
        i=0
        time.sleep(5)
        for img in imgs:
            if(i<=10):
                src = img.get_attribute('src')
                if src is not None:
                    if (src[0:4])=='http':
                        print(f'{product}_url : {src}')
                        i=i+1
            else:
                break
        driver_3.quit()
        time.sleep(10)
    
    except Exception as e:
        print(f"An error occurred: {e}")
        driver_3.quit()

products = ['fruits','cars','Machine Learning','Guitar','Cakes']

for product in products:
    search_img(product)
```

```python
# Q6 Write a program to scrap all the available details of best gaming laptops from digit.in.
def game():
    try:
        Name=[]
        O_S=[]
        D_size=[]
        Resolution=[]
        Processor=[]
        all_lists=['/html/body/div[1]/header/div/div[2]/div/nav/ul/li[2]/ul/li[3]/ul/li[1]/a','/html/body/div[1]/header/div/div[2]/div/nav/ul/li[2]/ul/li[3]/ul/li[2]/a','/html/body/div[1]/header/div/div[2]/div/nav/ul/li[2]/ul/li[3]/ul/li[3]/a','/html/body/div[1]/header/div/div[2]/div/nav/ul/li[2]/ul/li[3]/ul/li[4]/a']
        for li in all_lists:
            driver_6=webdriver.Chrome()
            driver_6.get('https://www.digit.in')
            driver_6.maximize_window()
            laptops = WebDriverWait(driver_6, 10).until(EC.presence_of_element_located((By.XPATH, '/html/body/div[1]/header/div/div[2]/div/nav/ul/li[2]/a')))
            actions = ActionChains(driver_6)
            actions.move_to_element(laptops).perform()
            best_gaming_laptops = WebDriverWait(driver_6, 10).until(EC.presence_of_element_located((By.XPATH, li)))
            best_gaming_laptops.click()
            time.sleep(5)
            name= WebDriverWait(driver_6, 10).until(EC.presence_of_all_elements_located((By.XPATH, '//h3[@class="font130 mt0 mb10 mobilesblockdisplay "]/a')))
            for n in name:
                Name.append(n.text)

            details = WebDriverWait(driver_6, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, '.woo_code_zone_loop')))        
            for div in details:
                spans = div.find_elements(By.CSS_SELECTOR, 'span.meta_v_value')

                span_texts = [span.text for span in spans]

                O_S.append(span_texts[0])
                D_size.append(span_texts[1])
                Resolution.append(span_texts[2])
                Processor.append(span_texts[3])
            time.sleep(5)
            driver_6.close()
                
        df=pd.DataFrame({'Name':Name,'Operating system':O_S,'Display Size':D_size,'Resolution':Resolution,'Processor':Processor})
        print(df)
            
        
            
    except Exception as e:
        print(f'An Error occured : ',e)
        
game()
```

```python
# Q8 Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted from any YouTube Video.
def extract_comments(driver):
    comments = []
    upvotes = []
    timestamps = []

    # Find all comment elements
    comment_elements = driver.find_elements(By.CSS_SELECTOR, "#content-text")
    upvote_elements = driver.find_elements(By.CSS_SELECTOR, "#vote-count-middle")
    time_elements = driver.find_elements(By.CSS_SELECTOR, "#published-time-text > a")
    

    for comment_element in comment_elements:
        comments.append(comment_element.text)
    for upvote_element in upvote_elements:
        upvotes.append(upvote_element.text)
    for time_element in time_elements:
        timestamps.append(time_element.text.strip())

    return comments, upvotes, timestamps

def collect_comments(video_url):
    driver = webdriver.Chrome()

    driver.get(video_url)
    time.sleep(2)

    body = driver.find_element(By.TAG_NAME, 'body')
    for _ in range(5):
        body.send_keys(Keys.PAGE_DOWN)
        time.sleep(1)

    comments, upvotes, timestamps = extract_comments(driver)

    while len(comments) < 500:
        body.send_keys(Keys.PAGE_DOWN)
        time.sleep(1)
        new_comments, new_upvotes, new_timestamps = extract_comments(driver)
        comments.extend(new_comments)
        upvotes.extend(new_upvotes)
        timestamps.extend(new_timestamps)

    return comments[:500], upvotes[:500], timestamps[:500]

video_url = "https://www.youtube.com/watch?v=OgRoRBLZbUQ"
comments, upvotes, timestamps = collect_comments(video_url)
time.sleep(5)

for c,u,t in zip(comments,upvotes,timestamps):
    print('comment :',c)
    print('upvote :',u)
    print('time :',t)
    print('##################################')

```

```python
''' Q9 Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” location. 
You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price,
dorms from price.'''
def hostel():
    try:
        driver_9=webdriver.Chrome()
        driver_9.get('https://www.hostelworld.com')
        time.sleep(5)
        loc_search=WebDriverWait(driver_9, 5).until( EC.presence_of_element_located((By.XPATH, '/html/body/div[3]/div/div[2]/main/header/div/div[2]/div[1]/div[1]/div/div[1]/div[1]/div/div[2]/label/input')))
        loc_search.send_keys('London')
        time.sleep(5)
        button_9= WebDriverWait(driver_9, 5).until( EC.presence_of_element_located((By.XPATH, '/html/body/div[3]/div/div[2]/main/header/div/div[2]/div[1]/div[1]/div/div[1]/div[2]/div/ul/li[2]/button')))
        button_9.click()
        details= WebDriverWait(driver_9, 5).until( EC.presence_of_element_located((By.XPATH, '/html/body/div[3]/div/div[2]/main/header/div/div[2]/div[1]/div[1]/div/div[5]/button[2]')))
        details.click()
        time.sleep(10)
        h_name=[]
        distance=[]
        rating=[]
        t_review=[]
        over_review=[]
        privates=[]
        dorms=[]
        for i in range(6,4,-1):
            names=driver_9.find_elements(By.XPATH,'//div[@class="property-name"]/span')
            for n in names:
                h_name.append(n.text)
            D=driver_9.find_elements(By.XPATH,'//span[@class="distance-description"]')
            for d in D:
                distance.append(d.text)


            ratings = driver_9.find_elements(By.XPATH, '//div[@class="property-info-header"]/div[2]/div/div[1]/span[1]')
            for r in ratings:
                rating.append(r.text)

            reviews = driver_9.find_elements(By.XPATH, '//div[@class="property-info-header"]/div[2]/div/div[1]/span[2]')
            for t in reviews:
                t_review.append(t.text)
                
            all_review=driver_9.find_elements(By.XPATH,'//div[@class="property-info-header"]/div[2]/div/div[2]/span')
            for o in all_review:
                over_review.append(o.text)

            private_prices=driver_9.find_elements(By.XPATH,'//div[@class="property-card"]/a/a/div[2]/div[4]/div[1]/div[1]')

            for price in private_prices:
                try:
                    private=price.find_element(By.TAG_NAME,'strong').text
                except NoSuchElementException:
                    private="-"
                privates.append(private)

            dorm_price=driver_9.find_elements(By.XPATH,'//div[@class="property-card"]/a/a/div[2]/div[4]/div[1]/div[2]')

            for price in dorm_price:
                try:
                    dorm=price.find_element(By.TAG_NAME,'strong').text
                except NoSuchElementException:
                    dorm="-"
                dorms.append(dorm)
                
            button_9 = WebDriverWait(driver_9, 10).until(
                EC.element_to_be_clickable((By.XPATH, f"/html/body/div[3]/div/div/div[2]/div[2]/div[1]/div/div/div[{i}]/div/button[2]"))
            )
            driver_9.execute_script("arguments[0].click();", button_9)
            
            
            time.sleep(10)
        for h,d,r,t,o,p,dorm in zip(h_name,distance,rating,t_review,over_review,privates,dorms):
  
            print('hostel Name: ',h)
            print('distance from city centre: ',d)
            print('rating',r)
            print('total review: ',t)
            print('overall review: ',o)
            print('privates from price: ',p)
            print('dorms from price : ',dorm)
            print('######################################')
        
    except TimeoutException:
        print("An error occurred: TimeoutException - The element could not be found within the given time frame.")
    except NoSuchElementException:
        print("An error occurred: NoSuchElementException - The element could not be found.")  
    except Exception as e:
        print('An Error Occur:',e)
        
hostel()
```
