```python

from bs4 import BeautifulSoup
import pandas as pd
import requests
import re
```

```python
# Q1 Write a python program to display IMDB’s Top rated 100 Indian movies’ data https://www.imdb.com/list/ls056092300/ (i.e. name, rating, year ofrelease) and make data frame.
names=[]
ratings=[]
year_ofrelease=[]
def imdb(url):
    res=requests.get(url)
    cont=BeautifulSoup(res.content)
    movie_data=cont.find_all(['div'],class_="lister-item-content")
    # print(movie_data)
    for i in movie_data:
        name = i.find('h3', class_='lister-item-header').find('a').text
        names.append(name)
        year=i.find('span',class_='lister-item-year text-muted unbold').text
        years = re.search(r'\b\d{4}\b', year)
        year_ofrelease.append(years.group())
        rating=i.find('span',class_='ipl-rating-star__rating').text
        ratings.append(rating)
    df=pd.DataFrame({'name':names,'rating':ratings,'year_of_release':year_ofrelease},index=range(1,101))
    print(df)
    

url='https://www.imdb.com/list/ls056092300'
imdb(url)
```

```python
# Q6.Write a python program to scrape first 10 product details which include product name , price , Image URL from https://www.bewakoof.com/bestseller?sort=popular .
def product(url):
    response=requests.get(url)
    content=BeautifulSoup(response.content)
    product_data=content.find_all('div',class_='plp-product-card')
    for i in product_data:
        name=i.find('h2',class_='clr-shade4').text
        price=i.find('div',class_='discountedPriceText').text
        print('product_name : ',name)
        print('price : ',price)
        if(i.find('div',class_='productImg')):
            image=i.find('div',class_='productImg').find('img')['src']
            print('image url : ',image)
        else:
            image=i.find('div',class_='productCardImg false').find('img')['src']
            print(image)
        print()
url='https://www.bewakoof.com/bestseller?sort=popular'
product(url)
```

```python
# Q7 .Please visit https://www.cnbc.com/world/?region=world and scrap- a) headings b) date c) News link
def visit(url):
    res=requests.get(url)
    cont=BeautifulSoup(res.content)
    datas=cont.find_all('div',class_='RiverPlusCard-cardLeft')
    for news in datas:
        heading=news.find('div',class_='RiverHeadline-headline RiverHeadline-hasThumbnail').find('a')
        dates=news.find('span',class_='RiverByline-datePublished')
        if(dates !=None):
            date=dates.text
        else:
            date=None
        if(heading.text==''):
            continue
        print('heading : ',heading.text)
        print('link : ',heading['href'])
        print('date : ',date)
        print('*--*'*30)
    
url='https://www.cnbc.com/world/?region=world'
visit(url)
```

```python
'''Q8 .Please visit https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/.
and scrap- a) Paper title b) date c) Author'''
def agriculture(url):
    res=requests.get(url)
    cont=BeautifulSoup(res.content)
    # print(cont)
    data=cont.find_all('div',class_='article-listing')
    for book in data:
        title=book.find('h2',class_='article-title').text.strip()
        date=book.find('p',class_='article-date').text
        author=book.find('p',class_='article-authors').text.strip()
        print('title : ',title)
        print('date : ', date)
        print('author :',author)
        print('*-'*50)
url='https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/'
agriculture(url)
```
