```python
# Q .2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.
def top_rated(url):
    
    # Fetch the HTML content of the webpage
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
    response = requests.get(url,headers=headers)
    content=BeautifulSoup(response.content)
    movie_name=content.find_all(['h3'],class_="ipc-title__text")
    remove_first=movie_name.pop(0)
    years=content.find_all('div',class_='sc-b189961a-7 feoqjK cli-title-metadata')
    # print(year[0].text.split()[0][:4])
    ratings=content.find_all('span',class_="ipc-rating-star ipc-rating-star--base ipc-rating-star--imdb ratingGroup--imdb-rating")
    # text = ratings[0].text.split()[0]
    name=[]
    rating=[]
    year=[]
    for i in range(0,100):
        name.append(movie_name[i].text[3:])
        year.append(years[i].text.split()[0][:4])
        rating.append(ratings[i].text.split()[0])
    return name,year,rating
    
imdb_url='https://www.imdb.com/chart/top'
name,year,rating=top_rated(imdb_url)
data=pd.DataFrame({'Name':name, 'Rating':rating, 'Year':year},index=range(1,101))
print(data)
```

```python
# Q.3) Write a python program to scrape mentioned details from dineout.co.in : i) Restaurant name ii) Cuisine iii) Location iv) Ratings v) Image URL.
def dineout(url):
    res=requests.get(url)
    cont=BeautifulSoup(res.content)
    restaurant = cont.find_all('div', class_='restnt-card restaurant')
    restaurant_names = []
    cuisines = []
    locations = []
    ratings = []
    image_urls = []
    for detail in restaurant:
        restaurant_name = detail.find('div', class_='restnt-info').find('a').text
        restaurant_names.append(restaurant_name)
        cuisine = detail.find('span', class_='double-line-ellipsis').text.split('|')
        cuisines.append(cuisine[1])
        location = detail.find('div', class_='restnt-info').find('div', class_='restnt-loc').text
        locations.append(location)
        rating = detail.find('div', class_='restnt-rating').text
        ratings.append(rating)
        image_url = detail.find('div', class_='img cursor').find('img')['data-src']
        image_urls.append(image_url)
        
    for i in range(len(restaurant_names)):
        print("Restaurant Name:", restaurant_names[i])
        print("Cuisine:", cuisines[i])
        print("Location:", locations[i])
        print("Rating:", ratings[i])
        print("Image URL:", image_urls[i])
        print()

dineurl='https://www.dineout.co.in/delhi-restaurants'
dineout(dineurl)
```

```python
# Q.4 Write a python program to display list of respected former finance minister of India(i.e. Name , Term of office) from https://presidentofindia.nic.in/former-presidents.htm and make data frame.
def president(url):
    res=requests.get(url)
    content=BeautifulSoup(res.content)
    data=content.find_all('div',class_='desc-sec')
    name=[]
    term=[]
    for i in data:
        name.append(i.find('h3').text)
        term.append(i.find('h5').text)
    return name,term   

pres='https://presidentofindia.nic.in/former-presidents'
name,term=president(pres)
finance_minister=pd.DataFrame({'Name':name,'Term of office':term},index=range(1,len(name)+1))
print(finance_minister)
```
